```python
import requests
import json
from datasets import load_dataset

corpus =  load_dataset("conll2003")
```

```python
sample = corpus['train']['tokens'][:20]
```

```python
sentences = [' '.join(sentence) for sentence in sample]
sentences[0]
```

```python
data = json.dumps({"signature_name": "serving_default", "data": sentences,"dataType":"text"})
headers = {"content-type": "application/json"}
json_response = requests.post(f'http://127.0.0.1:5001/predict/model_Nlp', data=data, headers=headers)
predictions = json_response.json()
print(predictions)
```
